````python
from agents import AsyncOpenAI, Agent, Runner, OpenAIChatCompletionsModel, RunConfig
from google.colab import userdata
from openai.types.responses import ResponseTextDeltaEvent
import asyncio

GEMINI_API_KEY = userdata.get('GEMINI_API_KEY')


external_client = AsyncOpenAI(
    api_key=GEMINI_API_KEY,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

modeler = OpenAIChatCompletionsModel(
    model="gemini-2.0-flash",
    openai_client=external_client
)

async def main():
  agent = Agent(
      name="Assistant",
      instructions="You are a helpful assistant.",
      model=modeler,
  )

  result = Runner.run_streamed(
      agent,
      "Please make 5 python beginner level quiz for me!"
  )
  async for event in result.stream_events():
      print(event)

asyncio.run(main())
````

Ab isko tod ke samjhte hain:

---

### **Line 1: `from openai.types.responses import ResponseTextDeltaEvent`**
- **Ye kya hai?**
  Ye line ek library se ek cheez (class ya type) ko import kar rahi hai. Library ka naam hai "openai", jo ek AI tool hai jo text bana sakta hai ya API ke zariye jawab de sakta hai.
- **Kya kaam kar rahi hai?**
  Ismein `ResponseTextDeltaEvent` naam ka ek tool ya type import ho raha hai, jo shayad OpenAI ke jawab ke chhote chhote hisson (deltas) ko handle karne ke liye hai. Matlab, jab AI text bana raha ho, to wo ek ek tukda ismein aayega.
- **Aasan zuban mein:**
  Ye line kehti hai ke hum ek AI library se ek khas cheez le rahe hain, jo text ke chhote tukdon ko samajhne mein madad karegi.

---

### **Line 2: `async for event in result.stream_events():`**
- **Ye kya hai?**
  Ye ek loop hai, jo "async" ka use karta hai. "Async" ka matlab hai ke ye code rukta nahi hai, balki cheezein jab taiyar hoti hain tab unko leta hai. `result` ek object hai (jo shayad pehle code mein banaya gaya hai), aur `stream_events()` uska ek function hai jo events ko ek ek karke deta hai.
- **Kya kaam kar rahi hai?**
  Ye AI se aane wale events (jaise text ke tukde) ko ek ek karke leta hai, jaise jaise wo aate hain. Ye "stream" ka matlab hai ke text thodi thodi der mein aata rahega, jaise live video ki tarah.
- **Aasan zuban mein:**
  Ye line kehti hai ke AI se jo cheezein (events) aayengi, hum unko ek ek karke check karenge, jaise jaise wo humein milengi.

---

### **Line 3: `if event.type == "raw_response_event" and isinstance(event.data, ResponseTextDeltaEvent):`**
- **Ye kya hai?**
  Ye ek condition hai jo check karti hai do cheezein:
  1. `event.type == "raw_response_event"`: Ye dekhta hai ke jo event aaya hai, uska type "raw_response_event" hai ya nahi.
  2. `isinstance(event.data, ResponseTextDeltaEvent)`: Ye check karta hai ke event ke andar jo data hai, wo `ResponseTextDeltaEvent` type ka hai ya nahi.
- **Kya kaam kar rahi hai?**
  Ye line sirf un events ko chunti hai jo khas tarah ke hain (raw response) aur jinka data text ke tukdon wala hai. Yani, sirf wahi events aage jayenge jo text ke chhote hisse la rahe hain.
- **Aasan zuban mein:**
  Ye kehta hai ke agar event ek khas tarah ka hai aur usmein text ka chhota tukda hai, to hi hum aage badhenge.

---

### **Line 4: Print Statement**
```python
print(event.data.delta, end="", flush=True)
```
- **Kya hai yeh?**: Yeh line screen pe kuch print karti hai, specifically `event.data.delta` ko.
- **Kaam kya karta hai?**:
  - **`event.data.delta`**: Yeh `event` ke `data` ke andar ek `delta` naam ka part hai, jo shayad text ka ek chhota update ya hissa hai jo API se aaya hai.
  - **`end=""`**: Normally, `print` ke baad ek nayi line shuru ho jati hai, lekin yeh kehta hai ke nayi line mat shuru karo, sab kuch ek hi line mein print karo.
  - **`flush=True`**: Yeh kehta hai ke jo bhi print ho raha hai, usko turant screen pe dikhao, wait mat karo. Yeh real-time output ke liye useful hai.
- **Simple bhasha mein**: Yeh line kehti hai, "Jo chhota sa text ka update aaya hai, usko turant screen pe dikhao, aur ek hi line mein rakho, alag-alag line mein mat bhejo."



````python
from agents import AsyncOpenAI, Agent, Runner, OpenAIChatCompletionsModel, RunConfig
from google.colab import userdata
from openai.types.responses import ResponseTextDeltaEvent
import asyncio

GEMINI_API_KEY = userdata.get('GEMINI_API_KEY')


external_client = AsyncOpenAI(
    api_key=GEMINI_API_KEY,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

modeler = OpenAIChatCompletionsModel(
    model="gemini-2.0-flash",
    openai_client=external_client
)

async def main():
  agent = Agent(
      name="Assistant",
      instructions="You are a helpful assistant.",
      model=modeler,
  )

  result = Runner.run_streamed(
      agent,
      "Please make 5 python beginner level quiz for me!"
  )
  async for event in result.stream_events():
      if event.type == "raw_response_event" and isinstance(event.data, ResponseTextDeltaEvent):
        print(event.data.delta, end="", flush=True)

asyncio.run(main())
````
