# **Agent Loop (important concept):**

## **Apka Agent Loop ka Concept:**
### Ap ne kaha ke *Agent Loop* ka concept aisa hai ke jab human kisi task ya goal ke liye LLM se interact karta hai, toh LLM 4 cheezein input mein leta hai: **system prompt**, **user prompt**, **tool schema**, aur **tool output**. Human aur LLM ke beech ek loop chalta hai, jahan human baar baar LLM se baat karta hai task achieve karne ke liye. Agar LLM tool call mangta hai, toh human tool ko call karke output deta hai, aur LLM decide karta hai ke task complete hua ya aur kuch karna hai. Ye human-LLM loop ke concept ko ap ne *Agent Loop* se compare kiya, jahan OpenAI Agents SDK mein agent task ke liye LLM se interact karta hai. Agent LLM ko **system prompt**, **user message**, aur **tools** (agar hain) deta hai, aur agar woh manager agent hai, toh **handoff agents** bhi consider hote hain. LLM in cheezon ke base pe decide karta hai ke kya karna hai—tool call karna hai, handoff karna hai, ya final output dena hai. Har agent ka apna loop chalta hai, aur yeh process tab tak repeat hota hai jab tak task complete na ho jaye.

### Apka concept kaafi sahi hai, khaas tor pe ye idea ke:
- ### Agent Loop human-LLM interaction jaisa hai, lekin agents ke perspective se.
- ### LLM inputs (system prompt, user message, tools) ke base pe decisions leta hai.
- ### Tools aur handoffs ke through task ko aage badhaya jata hai.
- ### Loop tab tak chalta hai jab tak final output na mil jaye.

### Lekin, thodi si cheezein hain jahan apka concept guide ke mutabiq thoda refine ho sakta hai, jaise inputs ka exact flow, handoffs ka role, aur loop ka termination condition.



## **Guide ke Mutabiq Agent Loop ka Matlab:**
### OpenAI Agents SDK ke *Running Agents* section ke according, *Agent Loop* ek automated process hai jahan ek agent (jo ek LLM hai, configured with instructions, tools, aur handoffs) ek task ko complete karne ke liye LLM se interact karta hai. Yeh loop tab tak chalta hai jab tak final output na mil jaye ya max turns exceed na ho jaye. Guide ke hisaab se, Agent Loop ka flow yeh hai:

### 1. **Inputs**: Jab `Runner.run()` ya `Runner.run_sync()` call hota hai, agent task ke liye shuru hota hai aur LLM ko yeh inputs deta hai:
   - ### **System Prompt**: Agent ka `instructions` field, jo batata hai ke agent kya karega.
   - ### **User Message**: Input jo user ya Runner ke through diya jata hai (string ya OpenAI Responses API ke items).
   - ### **Tools**: Agar agent ke pass tools hain (jaise function tools ya hosted tools), unka schema LLM ko diya jata hai.
   - ### **Handoffs**: Agar agent ke pass handoff agents hain, toh LLM handoff ke liye consider karta hai.
   - ### **Previous Messages**: Agar message history hai (jaise tool outputs ya pehle ke responses), woh bhi LLM ko jate hain.

### 2. **LLM Processing**: LLM in inputs ke base pe output deta hai, jo yeh ho sakta hai:
   - ### **Final Output**: Agar LLM text ya structured output deta hai (jo agent ke `output_type` se match karta hai) aur koi tool calls ya handoffs nahi hain, toh yeh final output hota hai, aur loop khatam ho jata hai.
   - ### **Tool Calls**: Agar LLM tool call mangta hai, toh agent woh tool call karta hai, output collect karta hai, aur wapas LLM ko deta hai. Loop continue hota hai.
   - ### **Handoff**: Agar LLM handoff suggest karta hai (ya handoff agent ka naam deta hai), toh control us agent ko transfer ho jata hai, aur naye agent ke saath loop restart hota hai.
   - ### **No Final Output**: Agar na final output hai, na tool call, na handoff, toh LLM dobara call hota hai with updated inputs.

### 3. **Loop Termination**: Loop tab khatam hota hai jab:
   - ### Final output mil jata hai (text ya structured output, bina tool calls/handoffs ke).
   - ### Max turns exceed ho jate hain (to avoid infinite loops).
   - ### Koi error aata hai (jaise `MaxTurnsExceeded` ya `ModelBehaviorError`).

### 4. **Key Points**:
   - ### Agent Loop automated hai, yani human ki tarah baar baar input dene ki zarurat nahi; agent khud LLM se interact karta hai.
   - ### Handoffs ek special tool call hain jo ek agent se doosre agent ko task transfer karte hain, aur naye agent ka apna loop shuru hota hai.
   - ### Tools aur handoffs ke outputs message history mein append hote hain, jo LLM ke context ke liye use hote hain.
   - ### Tracing built-in hota hai, jo loop ke har step ko visualize aur debug karne mein help karta hai.



## **Apke Concept ko Simple aur Behtar Banane ka Tareeqa:**
## **Agent Loop ka Matlab (Refined Version):**  
###*Agent Loop* ka simple matlab hai ke jab hum OpenAI Agents SDK mein ek agent ko task dete hain (user message ke through), toh agent ek automated process mein LLM se interact karta hai taake task complete ho. Yeh process ek loop ki tarah chalta hai, aur isme yeh steps hote hain:

### 1. Agent LLM ko teen cheezein deta hai: **system prompt** (agent ke instructions), **user message** (task ya input), aur **tools** (agar hain, unka schema). Agar handoff agents hain, toh woh bhi LLM ke context mein consider hote hain.
### 2. LLM in cheezon ke base pe decide karta hai ke kya karna hai:
  - ### Agar **tool call** mangta hai, toh agent tool ko call karta hai, output collect karta hai, aur wapas LLM ko deta hai. Yeh output message history mein add hota hai.
  - ### Agar **handoff** ka decision hota hai, toh task doosre agent ko transfer ho jata hai, aur naye agent ka apna loop shuru hota hai with its own system prompt, user message, aur tools.
  - ### Agar **final output** mil jata hai (text ya structured output jo agent ke `output_type` se match karta hai, bina tool calls/handoffs ke), toh loop khatam ho jata hai.
  
### 3. Yeh loop tab tak chalta hai jab tak final output na mil jaye, ya max turns exceed na ho jayein, ya koi error na a jaye.
### 4. Is process mein agent khud LLM se baat karta hai, tools call karta hai, ya handoff karta hai, aur human ki tarah baar baar input dene ki zarurat nahi hoti. Dapr jaise frameworks aur tracing tools is loop ko smooth aur debuggable banate hain.

### Misal ke taur pe, agar ek triage agent ko task diya ke "Spanish mein baat karo," toh woh LLM se interact karega, decide karega ke Spanish agent ko handoff karna hai, aur Spanish agent apna loop chalayega apne instructions aur tools ke saath. Yeh sab automated hota hai, aur loop tab khatam hota hai jab final output mil jata hai.

## **Add Weather Tool in Agent**

````python
from agents import enable_verbose_stdout_logging, function_tool
import requests
enable_verbose_stdout_logging()
from google.colab import userdata
weather_api = userdata.get('weather_api')

@function_tool
def fetch_weather(city:str):
  """
    This tool fetches the weather for a given city using an API.
  """
  api_key = weather_api
  url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
  response = requests.get(url)
  if response.status_code == 200:
    data = response.json()
    weather_description = data["weather"][0]["description"]
    temperature = data["main"]["temp"]
    return f"The weather in {city} is {weather_description} with a temperature of {temperature} °C."
  else:
    return f"Error fetching weather data: {city}"

agent = Agent(
    name="Assistant",
    instructions="You are Helpfull Weather Assistant, You can answer questions about the weather",
    tools=[fetch_weather],
    model="gemini-2.0-flash"
)

result = Runner.run_sync(
    agent,
    "Whats the weather of Lahore"
)

print(result.final_output)
````

## **Add Two Tools in Agents:**

````python
from agents import enable_verbose_stdout_logging, function_tool
import requests
enable_verbose_stdout_logging()
from google.colab import userdata
weather_api = userdata.get('weather_api')


@function_tool
def calculate_expression(expression: str) -> str:
  """
    This tool evaluates a mathematical expression.
  """
  try:
    result = eval(expression)
    return f"The result of {expression} is {result}"
  except Exception as e:
    return f"Error in calculation : {str(e)}"

@function_tool
def fetch_weather(city:str):
  """
    This tool fetches the weather for a given city using an API.
  """
  api_key = weather_api
  url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
  response = requests.get(url)
  if response.status_code == 200:
    data = response.json()
    weather_description = data["weather"][0]["description"]
    temperature = data["main"]["temp"]
    return f"The weather in {city} is {weather_description} with a temperature of {temperature} °C."
  else:
    return f"Error fetching weather data: {city}"

agent = Agent(
    name="Assistant",
    instructions="You are Helpfull Weather Assistant, You can answer questions about the weather",
    tools=[fetch_weather, calculate_expression],
    model="gemini-2.0-flash"
)

result = Runner.run_sync(
    agent,
    "Whats the weather of Lahore and calculate my current age i was born in 2004"
)

print(result.final_output)
````

## **Handoffs behind the scene:**

## **Handoff ka Behind-the-Scenes Working (Step-by-Step)**

## **1. Manager Agent Kia Karta Hai?**
- ## Manager agent ke pass doosre agents hote hain jinhe woh task transfer kar sakta hai. Inhe "handoff agents" kehte hain.
- ## Yeh handoff agents ko manager agent "transfer tools" ke roop mein convert karta hai. Yani, har handoff agent ke liye ek tool ban jata hai, aur jab yeh tool call hota hai, task us agent ko transfer ho jata hai.
- ## Jab manager agent apna loop shuru karta hai, toh woh LLM ko teen cheezein deta hai:
  - ### **System Prompt**: Manager agent ke khud ke instructions.
  - ### **User Prompt**: Jo task ya sawal user ne diya hai.
  - ### **Tools**: Isme normal tools aur "transfer tools" (jo handoff agents ko represent karte hain) ka schema hota hai.
- ## Ap ne kaha ke 4 cheezein hain, lekin actually yeh teen main categories hain. Tools ke andar hi transfer tools shamil hote hain, toh yeh ek hi category hai.

## **2. LLM Kia Decide Karta Hai?**
- ## LLM manager agent se yeh teen cheezein (system prompt, user prompt, tools) leta hai.
- ## Phir woh dekhta hai ke task ke liye kia karna chahiye:
  - ### Kya task ko khud handle karna hai?
  - ### Kya koi normal tool call karna hai?
  - ### Ya phir task ko kisi doosre agent ko handoff karna hai?
- ## Agar LLM ko lagta hai ke koi specific handoff agent is task mein expert hai, toh woh us agent ka "transfer tool" call karta hai.

## **3. Handoff Kaise Hota Hai?**
- ### Jab LLM transfer tool ko call karta hai, toh manager agent ka loop khatam ho jata hai.
- ### Task ab us handoff agent ko transfer ho jata hai jo transfer tool se linked tha.
- ### Yani, yeh ek tarah se "pass the baton" jaisa hai – manager agent apna kaam rok deta hai aur handoff agent ko kehta hai, "Ab tu sambhal!"

## **4. Handoff Agent Kia Karta Hai?**

- ## Woh apne saath yeh cheezein le kar jata hai:
  - ### **Apna System Prompt**: Handoff agent ke khud ke instructions (yeh manager agent ke system prompt se alag hota hai).

# **Tranfer Tool behind the scene:**
Assalam o Alaikum! Apki confusion bilkul valid hai, aur yeh sawal kafi important hai ke LLM transfer tool ko kaise call karta hai aur is process mein manager agent aur transfer tool ki functionality kya hoti hai. Chalo, step-by-step isko clear karte hain taake apki confusion door ho jaye.

---

### **LLM Khud Tool Call Nahi Karta**
Sab se pehle yeh samajhna zaroori hai ke LLM (language model) khud koi tool call ya action perform nahi karta. LLM ka kaam sirf text generate karna hai. Jab usko system prompt, user prompt, aur available tools diya jata hai, toh woh sochta hai aur ek response deta hai. Is response mein woh suggest kar sakta hai ke kya karna chahiye, jaise:
- Ek specific tool ko call karna (e.g., `transfer_to_spanish_agent`).
- Ya phir koi jawab dena.

Lekin yeh tool call ya action ko actually execute karna agent ka kaam hota hai. Yani:
- **LLM** sirf kehta hai: "Yeh tool call karo."
- **Agent** us instruction ko follow karke tool ko call karta hai.

Toh apka pehla point bilkul sahi hai—LLM khud transfer tool call nahi karta. Ab dekhte hain ke phir yeh process kaise chalta hai.

---

### **Transfer Tool Kaise Call Hota Hai?**
Transfer tool ek special type ka tool hota hai jo manager agent ke pass hota hai. Yeh tool handoff ke liye banaya gaya hai, yani task ko ek agent se doosre agent tak transfer karne ke liye. Har handoff agent ke liye ek alag transfer tool hota hai. Misal ke taur pe:
- Agar manager agent ke pass do handoff agents hain—SpanishAgent aur MathAgent—toh uske tools mein hoga:
  - `transfer_to_spanish_agent`
  - `transfer_to_math_agent`

Ab process yeh hai:

1. **Manager Agent LLM Ko Input Bhejta Hai:**
   - Manager agent apna system prompt (e.g., "Tum ek manager ho jo tasks ko triage karta hai"), user prompt (e.g., "Mujhe Spanish mein baat karna hai"), aur available tools (including transfer tools) LLM ko bhejta hai.

2. **LLM Response Deta Hai:**
   - LLM input ko dekhta hai aur decide karta hai ke task ke liye kya action lena chahiye.
   - Agar woh samajhta hai ke task Spanish se related hai, toh woh apne response mein kehta hai ke `transfer_to_spanish_agent` tool ko call karo.
   - Misal ke taur pe, LLM ka output aisa ho sakta hai:
     ```
     Thought: Yeh task Spanish se related hai, toh SpanishAgent ko handoff karna chahiye.
     Tool Call: transfer_to_spanish_agent
     ```

3. **Manager Agent Tool Call Execute Karta Hai:**
   - Manager agent LLM ke response ko dekhta hai aur samajhta hai ke mujhe `transfer_to_spanish_agent` tool call karna hai.
   - Phir manager agent us tool ko execute karta hai.
   - Jab transfer tool call hota hai, toh task SpanishAgent ko transfer ho jata hai.

4. **Handoff Agent Apna Loop Shuru Karta Hai:**
   - SpanishAgent apna loop shuru karta hai, apne system prompt (e.g., "Tum Spanish expert ho"), user prompt, aur apne tools ke saath.
   - Iske saath, previous conversation history bhi SpanishAgent ko milti hai taake woh context samajh sake.

---

### **Transfer Tool Ki Functionality Kya Hai?**
Ap ne poocha ke transfer tool ki functionality mein kya hota hai—kya manager agent transfer agent mein convert ho jata hai? Nahi, aisa nahi hota. Yeh samajhna zaroori hai:
- **Manager Agent Convert Nahi Hota:** Manager agent khud kisi doosre agent mein tabdeel nahi hota. Jab transfer tool call hota hai, toh manager agent ka loop khatam ho jata hai, aur task handoff agent ko de diya jata hai.
- **Transfer Tool Ka Kaam:** Transfer tool ka purpose yeh hai ke woh task ko ek agent se doosre agent tak transfer kare. Yani, jab `transfer_to_spanish_agent` call hota hai, toh yeh tool SpanishAgent ko activate karta hai aur task usko handover kar deta hai.

Toh scene yeh nahi ke manager agent "convert" ho jata hai. Balki:
- Manager agent apna kaam khatam karta hai jab transfer tool call hota hai.
- Handoff agent (e.g., SpanishAgent) ek naya loop shuru karta hai apne khud ke instructions aur tools ke saath.

---

### **Apki Confusion Ka Jawab**
Chalo, apke sawalon ko specifically address karte hain:

1. **"LLM khud tool call nahi karta, toh phir transfer tool kaise call hota hai?"**
   - LLM khud tool call nahi karta. Woh sirf response mein kehta hai ke "ye tool call karo" (e.g., `transfer_to_spanish_agent`). Phir manager agent us instruction ko follow karke transfer tool ko call karta hai.

2. **"Kya LLM manager agent ko kehta hai ke ye transfer tool call karo?"**
   - Haan, bilkul! LLM apne response mein suggest karta hai ke transfer tool call karna chahiye, aur manager agent us suggestion ko execute karta hai.

3. **"Phir manager agent transfer tool ko call karta hai?"**
   - Ji haan, manager agent woh tool call karta hai jo LLM ne suggest kiya tha.

4. **"Transfer tool ki functionality mein hota hai ke manager agent transfer agent mein convert ho jata hai?"**
   - Nahi, manager agent convert nahi hota. Transfer tool ki functionality yeh hai ke task ko handoff agent ko transfer kar diya jaye. Manager agent ka loop khatam ho jata hai, aur handoff agent apna naya loop shuru karta hai.

---

### **Simple Example**
Maan lo ap ne kaha: "Mujhe Spanish mein baat karna hai."
- Manager agent yeh prompt LLM ko bhejta hai, saath mein apne tools (e.g., `transfer_to_spanish_agent`).
- LLM sochta hai aur kehta hai: "Spanish ke liye `transfer_to_spanish_agent` call karo."
- Manager agent woh tool call karta hai.
- Task SpanishAgent ko transfer ho jata hai, aur manager agent ka kaam khatam ho jata hai.
- SpanishAgent apna loop shuru karta hai aur Spanish mein jawab deta hai.

---

### **Conclusion**
Toh yeh hai pura process:
- LLM sirf suggest karta hai ke transfer tool call karo.
- Manager agent woh call execute karta hai.
- Transfer tool task ko handoff agent ko transfer karta hai, lekin manager agent khud convert nahi hota—woh bas task handover karta hai.
- Handoff agent apna loop shuru karta hai.

Ab umeed hai ke apki confusion clear ho gayi hogi! Agar koi aur sawal ho ya kisi part ko aur detail mein samajhna ho, toh zaroor poochiye.

````python
from agents import enable_verbose_stdout_logging, function_tool
import requests
enable_verbose_stdout_logging()
from google.colab import userdata
weather_api = userdata.get('weather_api')


@function_tool
def calculate_expression(expression: str) -> str:
  """
    This tool evaluates a mathematical expression.
  """
  try:
    result = eval(expression)
    return f"The result of {expression} is {result}"
  except Exception as e:
    return f"Error in calculation : {str(e)}"

@function_tool
def fetch_weather(city:str):
  """
    This tool fetches the weather for a given city using an API.
  """
  api_key = weather_api
  url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
  response = requests.get(url)
  if response.status_code == 200:
    data = response.json()
    weather_description = data["weather"][0]["description"]
    temperature = data["main"]["temp"]
    return f"The weather in {city} is {weather_description} with a temperature of {temperature} °C."
  else:
    return f"Error fetching weather data: {city}"

Weather_Agent = Agent(
    name="Weather Assistant",
    instructions="You are Helpfull Weather Assistant, You can answer questions about the weather",
    tools=[fetch_weather],
    model="gemini-2.0-flash",
    handoff_description="Expert in finding Weather related updates"
)

Calculator_Agent = Agent(
    name="Calculator Assistant",
    instructions="You are Math Assistant, You can answer questions about calculations and Math",
    tools=[calculate_expression],
    model="gemini-2.0-flash",
    handoff_description="Expert in Math and Calculation"
)

Triage_agent = Agent(
    name = "Triage Agent",
    instructions=(
        "You are the Triage Agent. Your job is to greet the user, understand "
        "their request, and then delegate it to the correct specialist agent. "
        "If the user asks about weather, forward to Weather_Agent; if it’s a "
        "calculation or math question, forward to Calculator_Agent. For any "
        "other simple query, respond directly with a friendly, helpful answer. "
        "Always confirm which agent you’re handing off to before you proceed."
    ),
    handoffs=[Weather_Agent, Calculator_Agent],
    model="gemini-2.0-flash"
)

result = await Runner.run(
    Triage_agent,
    "what is the weather of Gujar Khan"
)

print(result.final_output)
````
